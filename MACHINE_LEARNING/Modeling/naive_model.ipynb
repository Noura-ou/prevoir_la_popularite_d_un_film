### CatBoostEncoder()

Le CatBoostEncoder est un encodeur de caractéristiques (feature encoder) utilisé dans le domaine du Machine Learning pour traiter des données catégorielles. Cet encodeur est spécifiquement conçu pour fonctionner avec l'algorithme de gradient boosting CatBoost, mais il peut également être utilisé avec d'autres modèles.

Le CatBoostEncoder utilise une technique d'encodage basée sur les cibles pour convertir les variables catégorielles en valeurs numériques. 

Exemple:

```python
# Define catboost encoder
cbe_encoder = ce.cat_boost.CatBoostEncoder()
  
# Fit encoder and transform the features
cbe_encoder.fit(X, y)
X_cbe = cbe_encoder.transform(X)
print(X_cbe)
```
