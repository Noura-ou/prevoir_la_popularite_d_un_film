```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger votre DataFrame "df"
df = pd.read_csv("../Feature_engineering/data/dataset_model_silver.csv")


plt.figure(figsize=(10, 6))
top_10_genres = df['genre'].value_counts().nlargest(10)
sns.barplot(x=top_10_genres.index, y=top_10_genres.values)
plt.title("Top 10 genres avec le nombre de films")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Genre")
plt.ylabel("Nombre de films")
plt.show()



```

```python
plt.figure(figsize=(8, 8))
top_10_nationalities = df['langue_d_origine'].value_counts().nlargest(3)
top_10_nationalities.plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 3 des langues les plus presentes avec leur répartition")
plt.ylabel('')
plt.show()
```

```python
# Histogramme pour la durée des films
plt.figure(figsize=(10, 6))
sns.histplot(df['duree'].dropna(), bins=20)
plt.title("Distribution des durées des films")
plt.xlabel("Durée (minutes)")
plt.ylabel("Nombre de films")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10_articles = df.nlargest(8, 'nombre_article')
sns.barplot(x='nombre_article', y='titre', data=top_10_articles)
plt.title("Top 5 films avec le plus grand nombre d'articles de presse")
plt.xlabel("Nombre d'articles")
plt.ylabel("Titre du film")
plt.show()
```
