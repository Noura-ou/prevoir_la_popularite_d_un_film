## <font color="green"> || - Data Analyse</font>

### 1 ) Histogramme de la durée des films :

```python
import matplotlib.pyplot as plt

plt.hist(data['durée'], bins=20, color='skyblue')
plt.xlabel('Durée (en minutes)')
plt.ylabel('Nombre de films')
plt.title('Distribution de la durée des films')
plt.show()
```

### 2) Nuage de points pour visualiser la corrélation entre les notes de presse et les notes spectateurs

```python
plt.scatter(data['note_presse'], data['note_spectateurs'], alpha=0.5)
plt.xlabel('Note de presse')
plt.ylabel('Note spectateurs')
plt.title('Corrélation entre les notes de presse et les notes spectateurs')
plt.show()
```

### 3) Diagramme en secteurs pour visualiser la répartition des types de films :

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='type_film', palette='pastel')
plt.xlabel('Type de film')
plt.ylabel('Nombre de films')
plt.title('Répartition des types de films')
plt.xticks(rotation=45, ha='right')
plt.show()
```
