```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger votre DataFrame "df"
df = pd.read_csv("dataset_silver.csv")


plt.figure(figsize=(10, 6))
top_10_genres = df['genre'].value_counts().nlargest(10)
sns.barplot(x=top_10_genres.index, y=top_10_genres.values)
plt.title("Top 10 genres avec le nombre de films")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Genre")
plt.ylabel("Nombre de films")
plt.show()



```

```python
plt.figure(figsize=(8, 8))
top_10_nationalities = df['langue_d_origine'].value_counts().nlargest(3)
top_10_nationalities.plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 3 des langues les plus presentes avec leur répartition")
plt.ylabel('')
plt.show()
```

```python
# Histogramme pour la durée des films
plt.figure(figsize=(10, 6))
sns.histplot(df['durée'].dropna(), bins=20)
plt.title("Distribution des durées des films")
plt.xlabel("Durée (minutes)")
plt.ylabel("Nombre de films")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10_articles = df.nlargest(8, 'nombre_article')
sns.barplot(x='nombre_article', y='titre', data=top_10_articles)
plt.title("Top 5 films avec le plus grand nombre d'articles de presse")
plt.xlabel("Nombre d'articles")
plt.ylabel("Titre du film")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10_box_office = df.nlargest(10, 'box_office_total')
sns.barplot(x='box_office_total', y='titre', data=top_10_box_office, ci=None)
plt.title("Top 10 films avec les plus grandes recettes au box-office")
plt.xlabel("Recettes au box-office")
plt.ylabel("Titre du film")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10_notes = df.nlargest(6, ['note_presse', 'note_spectateurs'])

# Créer un barplot pour la note de presse
sns.barplot(x='note_presse', y='titre', data=top_10_notes, color='blue', label='Note presse')

# Créer un barplot pour la note des spectateurs
sns.barplot(x='note_spectateurs', y='titre', data=top_10_notes, color='orange', label='Note spectateurs')

# Afficher la légende
plt.legend()

plt.title("Top 5 films avec les meilleures notes de presse et de spectateurs")
plt.xlabel("Note")
plt.ylabel("Titre du film")
plt.show()
```
