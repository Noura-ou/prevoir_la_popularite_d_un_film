# Merging the datasets

```python
import pandas as pd
```

## Import CSV

```python
boxoffice_tous= pd.read_csv('../SCRAP_DATA_FOR_THE_MODEL/dataset/boxoffice.csv', dtype={'boxoffice': pd.Int64Dtype()})

merged_boxoffice = boxoffice_tous
```

```python
movies_tous = pd.read_csv('../SCRAP_DATA_FOR_THE_MODEL/dataset/movies.csv')

merged_movies = movies_tous
```

## Movies dataset

### Supprimer les doublons

```python
merged_movies = merged_movies.drop_duplicates()
```

## Merging the datasets boxoffice and movies

```python
merged_dataset = pd.merge(merged_movies, merged_boxoffice, on='film_id_allocine')
```

```python
merged_dataset.info()
```

```python
merged_dataset.head()
```

## Choisir le bon boxoffice

```python
# Comparer les colonnes "date" et "fin_semaine_1" pour déterminer quelle colonne "boxoffice" prendre
mask = merged_dataset['date'] == merged_dataset['fin_semaine_1']
merged_dataset['boxoffice'] = merged_dataset['boxoffice_2'].where(mask, merged_dataset['boxoffice_1'])

# Supprimer les colonnes inutiles
merged_dataset.drop(columns=['boxoffice_1', 'boxoffice_2', 'fin_semaine_1'], inplace=True)

# Afficher le dataframe final
print(merged_dataset)
```

```python
merged_dataset.info()
```

```python
default_date = pd.to_datetime('1970-01-01')
merged_dataset['date'] = merged_dataset['date'].apply(lambda x: default_date if x is None else x)
```

```python
merged_dataset['annee_production'] = merged_dataset['annee_production'].astype(int)
merged_dataset['film_id_allocine'] = merged_dataset['film_id_allocine'].astype(int)
merged_dataset['boxoffice'] = merged_dataset['boxoffice'].astype(int)
```

```python
import re
incorrect_dates = merged_dataset[~merged_dataset['date'].str.match(r'\d{2}/\d{2}/\d{4}')]
incorrect_dates
```

```python
merged_dataset = merged_dataset[merged_dataset['date'].str.match(r'\d{2}/\d{2}/\d{4}')]

merged_dataset['date'] = pd.to_datetime(merged_dataset['date'], format='%d/%m/%Y')

merged_dataset = merged_dataset.sort_values(by='date')

date_de_separation = pd.to_datetime('2023-07-01')  # Remplacez cette date par celle que vous souhaitez

# Diviser le dataset en deux parties
dataset_model = merged_dataset[merged_dataset['date'] < date_de_separation]
dataset_history_django = merged_dataset[merged_dataset['date'] > date_de_separation]

dataset_model['date'] = dataset_model['date'].replace(default_date, pd.NA)
```

```python
dataset_history_django.head(10)
```

```python
dataset_model.info()
```

```python
dataset_history_django_boxoffice = dataset_history_django[['film_id_allocine', 'boxoffice']]
dataset_history_django_features = dataset_history_django.drop(columns=['boxoffice'])
```

```python
def extract_first_language(langue_d_origine_str):
    languages = langue_d_origine_str.split(',')
    return languages[0].strip() if languages else None

# Appliquer la fonction sur la colonne "langue_d_origine" pour extraire la première langue d'origine
dataset_model['langue_d_origine'] = dataset_model['langue_d_origine'].apply(extract_first_language)
```

```python
dataset_model.info()
```

```python
dataset_history_django_features['genre'] = dataset_history_django_features['genre'].apply(lambda x: x.split('_')[0])
```

```python

from datetime import  datetime # Obtenir la date d'aujourd'hui
today = pd.Timestamp(datetime.now().date())

# Filtrer les lignes avec une date antérieure ou égale à aujourd'hui
dataset_history_django_features = dataset_history_django_features[dataset_history_django_features['date'] <= today]
```
